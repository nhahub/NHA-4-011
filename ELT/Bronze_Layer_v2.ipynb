```python
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
import pandas as pd
```

# Data Loading

```python
spark = SparkSession.builder.getOrCreate()
```

```python
df = spark.read.format('csv')\
    .option("header", "true")\
    .option("inferSchema", "true")\
    .load("Files/data_depi_final.csv")
```

```python
display(df)
```

# Column Renaming

```python
df.columns
```

```python
clean_columns = [
    'Student_ID',
    'First_Name',
    'Last_Name',
    'Email',
    'Gender',
    'Department',
    'Grade',
    'Extracurricular_Activities',
    'Internet_Access_at_Home',
    'Parent_Education_Level',
    'Family_Income_Level',
    'Age',
    'Math_Score',
    'assi_Math',
    'Science_Score',
    'assi_Science',
    'English_Score',
    'assi_English',
    'Social_Studies_Score',
    'assi_Social_Studie',
    'French_Score',
    'assi_French',
    'Computer_Science_Score',
    'assi_Computer_Science',
    'last_score',
    'Study_Hours_per_Week',
    'Stress_Level_1_10',
    'Sleep_Hours_per_Night',
    'assi_late',
    'non_academic_attendance',
    'academic_attendance',
    'lms_login_freq_per_day',
    'lms_active_avg_hrs',
    'resource_access_avg_hrs'
]
```

```python
old_cols = df.columns

for old, new in zip(old_cols, clean_columns):
    df = df.withColumnRenamed(old, new)
```

# Save to Delta

```python
df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze.students")
```
