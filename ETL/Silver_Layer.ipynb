```python
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.functions import col, when, rand, round as spark_round, isnull
import pandas as pd
```

```python
df.filter(
    (col("Student_ID").isNull()) |
    (col("First_Name").isNull()) |
    (col("Last_Name").isNull()) |
    (col("Email").isNull()) |
    (col("Gender").isNull()) |
    (col("Department").isNull()) |
    (col("Grade").isNull()) |
    (col("Extracurricular_Activities").isNull()) |
    (col("Internet_Access_at_Home").isNull()) |
    (col("Parent_Education_Level").isNull()) |
    (col("Family_Income_Level").isNull()) |
    (col("Age").isNull()) |
    (col("Math_Score").isNull()) |
    (col("assi_Math").isNull()) |
    (col("Science_Score").isNull()) |
    (col("assi_Science").isNull()) |
    (col("English_Score").isNull()) |
    (col("assi_English").isNull()) |
    (col("Social_Studies_Score").isNull()) |
    (col("assi_Social_Studie").isNull()) |
    (col("French_Score").isNull()) |
    (col("assi_French").isNull()) |
    (col("Computer_Science_Score").isNull()) |
    (col("assi_Computer_Science").isNull()) |
    (col("last_score").isNull()) |
    (col("Study_Hours_per_Week").isNull()) |
    (col("Stress_Level_1_10").isNull()) |
    (col("Sleep_Hours_per_Night").isNull()) |
    (col("assi_late").isNull()) |
    (col("non_academic_attendance").isNull()) |
    (col("academic_attendance").isNull()) |
    (col("lms_login_freq_per_day").isNull()) |
    (col("lms_active_avg_hrs").isNull()) |
    (col("resource_access_avg_hrs").isNull())
).toPandas()
```

```python
null_counts = df.select([
    sum(when(col(c).isNull(), 1).otherwise(0)).alias(c)
    for c in df.columns
])

display(null_counts)
```

```python
df = df.withColumn("Total_Avg", 
    (col("Math_Score") + col("Science_Score") + col("English_Score") + 
     col("Social_Studies_Score") + col("French_Score") + col("Computer_Science_Score")) / 6
)
```

```python
df = df.withColumn("Internet_Access_at_Home",
    when(col("Family_Income_Level") == "High", when(rand() <= 0.95, "Yes").otherwise("No"))
    .when(col("Family_Income_Level") == "Medium", when(rand() <= 0.75, "Yes").otherwise("No"))
    .when(col("Family_Income_Level") == "Low", when(rand() <= 0.35, "Yes").otherwise("No"))
    .otherwise(col("Internet_Access_at_Home"))
)
```

```python
df = df.withColumn("academic_attendance",
    when(isnull(col("academic_attendance")),
        when(col("Total_Avg") >= 85, spark_round(rand() * (100 - 90) + 90, 2))
        .when(col("Total_Avg") >= 70, spark_round(rand() * (89 - 75) + 75, 2))
        .otherwise(spark_round(rand() * (70 - 50) + 50, 2))
    ).otherwise(col("academic_attendance"))
)
```

```python
df = df.withColumn("Math_Score",
    when(col("Department") == "Middle School", col("Math_Score") + 12 + (rand() * 6 - 3))
    .when(col("Department") == "6th Grade", col("Math_Score") + 2 + (rand() * 6 - 3))
    .when(col("Department") == "5th Grade", col("Math_Score") - 8 + (rand() * 6 - 3))
    .otherwise(col("Math_Score"))
)
```

```python
df = df.withColumn("Math_Score", 
    when(col("Math_Score") > 100, 100)
    .when(col("Math_Score") < 0, 0)
    .otherwise(spark_round(col("Math_Score"), 2))
)
```

```python
df = df.withColumn("Grade",
    when(col("Total_Avg") >= 88, "A")
    .when(col("Total_Avg") >= 75, "B")
    .when(col("Total_Avg") >= 65, "C")
    .when(col("Total_Avg") >= 50, "D")
    .otherwise("F")
)
```

```python
df = df.withColumn("lms_active_avg_hrs",
    when(col("Internet_Access_at_Home") == "No", spark_round(col("lms_active_avg_hrs") * 0.4, 2))
    .otherwise(spark_round(col("lms_active_avg_hrs") * 1.2, 2))
)
```

```python
df = df.withColumn("resource_access_avg_hrs",
    when(col("Internet_Access_at_Home") == "No", spark_round(col("resource_access_avg_hrs") * 0.3, 2))
    .otherwise(spark_round(col("resource_access_avg_hrs") * 1.2, 2))
)
```

```python
df = df.withColumn("Study_Hours_per_Week",
    when(col("Grade") == "A", spark_round(rand() * (35 - 26) + 26, 1))
    .when(col("Grade") == "B", spark_round(rand() * (26 - 18) + 18, 1))
    .when(col("Grade") == "C", spark_round(rand() * (18 - 12) + 12, 1))
    .when(col("Grade") == "D", spark_round(rand() * (12 - 6) + 6, 1))
    .otherwise(spark_round(rand() * (7 - 2) + 2, 1)) # للراسبين F
)
```

```python
df = df.withColumn("Extracurricular_Activities",
    when(col("Grade") == "A", when(rand() <= 0.75, "Yes").otherwise("No"))
    .when(col("Grade") == "B", when(rand() <= 0.55, "Yes").otherwise("No"))
    .when(col("Grade") == "C", when(rand() <= 0.35, "Yes").otherwise("No"))
    .when(col("Grade") == "D", when(rand() <= 0.20, "Yes").otherwise("No"))
    .otherwise(when(rand() <= 0.10, "Yes").otherwise("No"))
)
```

```python
df = df.withColumn("non-academic_attendance",
    when(col("Extracurricular_Activities") == "Yes", spark_round(rand() * (100 - 75) + 75, 2))
    .otherwise(spark_round(rand() * (65 - 15) + 15, 2))
)
```

```python
df = df.withColumn("academic_attendance",
    when(col("Grade") == "A", spark_round(rand() * (100 - 90) + 90, 2))
    .when(col("Grade") == "B", spark_round(rand() * (95 - 80) + 80, 2))
    .when(col("Grade") == "C", spark_round(rand() * (85 - 70) + 70, 2))
    .when(col("Grade") == "D", spark_round(rand() * (75 - 50) + 50, 2))
    .otherwise(spark_round(rand() * (55 - 20) + 20, 2))
)
```

```python
df = df.withColumn("lms_login_freq_per_day",
    when(col("Grade") == "A", spark_round(rand() * (3.0 - 2.5) + 2.5, 2))
    .when(col("Grade") == "B", spark_round(rand() * (2.5 - 1.8) + 1.8, 2))
    .when(col("Grade") == "C", spark_round(rand() * (1.8 - 1.2) + 1.2, 2))
    .when(col("Grade") == "D", spark_round(rand() * (1.2 - 0.6) + 0.6, 2))
    .otherwise(spark_round(rand() * (0.6 - 0.1) + 0.1, 2))
)
```

```python
df = df.withColumn("resource_access_avg_hrs",
    when(col("Grade") == "A", spark_round(rand() * (2.8 - 2.0) + 2.0, 2))
    .when(col("Grade") == "B", spark_round(rand() * (2.0 - 1.4) + 1.4, 2))
    .when(col("Grade") == "C", spark_round(rand() * (1.4 - 0.9) + 0.9, 2))
    .when(col("Grade") == "D", spark_round(rand() * (0.9 - 0.4) + 0.4, 2))
    .otherwise(spark_round(rand() * (0.4 - 0.1) + 0.1, 2))
)
```

```python
df = df.withColumn("Sleep_Hours_per_Night",
    when(col("Grade") == "A", spark_round(rand() * (9.0 - 7.5) + 7.5, 1))
    .when(col("Grade") == "B", spark_round(rand() * (8.0 - 6.5) + 6.5, 1))
    .when(col("Grade") == "C", spark_round(rand() * (7.0 - 5.5) + 5.5, 1))
    .when(col("Grade") == "D", spark_round(rand() * (6.0 - 4.5) + 4.5, 1))
    .otherwise(spark_round(rand() * (5.0 - 4.0) + 4.0, 1))
)
```

```python
df = df.withColumn("Stress_Level_Calc", 
    10 - ((col("Sleep_Hours_per_Night") - 4) / 5) * 8 + (rand() * 2 - 1)
)
```

```python
df = df.withColumn("Stress_Level (1-10)",
    when(spark_round(col("Stress_Level_Calc")) > 10, 10)
    .when(spark_round(col("Stress_Level_Calc")) < 1, 1)
    .otherwise(spark_round(col("Stress_Level_Calc")).cast("integer"))
).drop("Stress_Level_Calc") 
```

```python
df = df.withColumn("assi_late",
    when(col("Stress_Level (1-10)") >= 8, spark_round(rand() * (5 - 2) + 2, 0).cast("integer"))
    .when(col("Stress_Level (1-10)") >= 5, spark_round(rand() * (3 - 1) + 1, 0).cast("integer"))
    .otherwise(spark_round(rand() * (2 - 0) + 0, 0).cast("integer"))
)
```

```python
df = df.withColumn("assi_late", ceil(col("assi_late")))
```

```python
df = df.withColumn("id", monotonically_increasing_id())
```

```python
df = df.select("id", *[c for c in df.columns if c != 'id'])
display(df)
```

```python
df = df.drop(
    "non_academic_attendance",
    "Stress_Level_1_10"
)
```

```python
df = df.withColumn("Math_Score", round(col("Math_Score"), 2)) \
    .withColumn("Science_Score", round(col("Science_Score"), 2)) \
    .withColumn("English_Score", round(col("English_Score"), 2)) \
    .withColumn("Social_Studies_Score", round(col("Social_Studies_Score"), 2)) \
    .withColumn("French_Score", round(col("French_Score"), 2)) \
    .withColumn("Computer_Science_Score", round(col("Computer_Science_Score"), 2)) \
    .withColumn("assi_Math", round(col("assi_Math"), 2)) \
    .withColumn("assi_Science", round(col("assi_Science"), 2)) \
    .withColumn("assi_English", round(col("assi_English"), 2)) \
    .withColumn("assi_Social_Studie", round(col("assi_Social_Studie"), 2)) \
    .withColumn("assi_French", round(col("assi_French"), 2)) \
    .withColumn("assi_Computer_Science", round(col("assi_Computer_Science"), 2)) \
    .withColumn("last_score", round(col("last_score"), 2)) \
    .withColumn("Study_Hours_per_Week", round(col("Study_Hours_per_Week"), 2)) \
    .withColumn("Stress_Level_1_10", round(col("Stress_Level_1_10"), 2)) \
    .withColumn("Sleep_Hours_per_Night", round(col("Sleep_Hours_per_Night"), 2)) \
    .withColumn("assi_late", round(col("assi_late"), 2)) \
    .withColumn("non_academic_attendance", round(col("non_academic_attendance"), 2)) \
    .withColumn("academic_attendance", round(col("academic_attendance"), 2)) \
    .withColumn("lms_login_freq_per_day", round(col("lms_login_freq_per_day"), 2)) \
    .withColumn("lms_active_avg_hrs", round(col("lms_active_avg_hrs"), 2)) \
    .withColumn("resource_access_avg_hrs", round(col("resource_access_avg_hrs"), 2))
```

```python
result = df.groupBy("Gender", "Grade", "Department") \
  .agg(
      avg("assi_late").alias("avg_assi_late"),
      avg("non_academic_attendance").alias("avg_non_academic"),
      avg("academic_attendance").alias("avg_academic"),
      avg("lms_login_freq_per_day").alias("avg_login"),
      avg("lms_active_avg_hrs").alias("avg_lms_hours"),
      avg("resource_access_avg_hrs").alias("avg_resource")
  ) \
  .orderBy("Grade", "Department")

display(result)
```
