```sql
CREATE TABLE Dim_Identity (
    id INT NOT NULL,
    First_Name STRING,
    Last_Name STRING,
    Email STRING,
    Gender STRING,
    Age INT
);

CREATE TABLE Dim_Academic_Context (
    id INT NOT NULL,
    Department STRING,
    Grade STRING,
    Extracurricular_Activities STRING
);

CREATE TABLE Dim_Social (
    id INT NOT NULL,
    Internet_Access_at_Home STRING,
    Parent_Education_Level STRING,
    Family_Income_Level STRING
);

CREATE TABLE Dim_Assignments (
    id INT NOT NULL,
    assi_Math STRING,
    assi_Science STRING,
    assi_English STRING,
    assi_Social_Studie STRING,
    assi_French STRING,
    assi_Computer_Science STRING,
    assi_late STRING
);


CREATE TABLE Fact_Performance (
    id INT NOT NULL,
    Student_ID STRING,
    Math_Score FLOAT,
    Science_Score FLOAT,
    English_Score FLOAT,
    Social_Studies_Score FLOAT,
    French_Score FLOAT,
    Computer_Science_Score FLOAT,
    last_score FLOAT,
    Total_Avg FLOAT,
    Study_Hours_per_Week FLOAT,
    Stress_Level INT,
    Stress_Level_1_10 INT,
    Sleep_Hours_per_Night FLOAT,
    academic_attendance FLOAT,
    non_academic_attendance FLOAT,
    lms_login_freq_per_day FLOAT,
    lms_active_avg_hrs FLOAT,
    resource_access_avg_hrs FLOAT
);
```

```sql
CREATE OR REPLACE VIEW gold.academic_demographics_analysis AS
SELECT 
    Department, Grade, Gender, Age,
    Math_Score, Science_Score, English_Score, Social_Studies_Score, 
    French_Score, Computer_Science_Score, Total_Avg
FROM Fact_Performance f
JOIN Dim_Identity i ON f.id = i.id
JOIN Dim_Academic_Context a ON f.id = a.id;

CREATE OR REPLACE VIEW gold.socioeconomic_impact AS
SELECT 
    s.Parent_Education_Level, s.Family_Income_Level, s.Internet_Access_at_Home,
    a.Grade, f.lms_active_avg_hrs, f.resource_access_avg_hrs
FROM Dim_Social s
JOIN Dim_Academic_Context a ON s.id = a.id
JOIN Fact_Performance f ON s.id = f.id;

CREATE OR REPLACE VIEW gold.study_habits_stress_analysis AS
SELECT 
    f.Study_Hours_per_Week, f.Sleep_Hours_per_Night, f.Stress_Level_1_10,
    a.Grade, a.Extracurricular_Activities
FROM Fact_Performance f
JOIN Dim_Academic_Context a ON f.id = a.id;

CREATE OR REPLACE VIEW gold.lms_engagement_analysis AS
SELECT 
    f.lms_login_freq_per_day, f.lms_active_avg_hrs, f.resource_access_avg_hrs,
    f.Math_Score, f.Science_Score, a.Grade
FROM Fact_Performance f
JOIN Dim_Academic_Context a ON f.id = a.id;

CREATE OR REPLACE VIEW gold.attendance_and_submission_behavior AS
SELECT 
    f.academic_attendance, f.non_academic_attendance, f.Stress_Level_1_10, f.lms_login_freq_per_day,
    a.Grade, a.Extracurricular_Activities,
    d.assi_late
FROM Fact_Performance f
JOIN Dim_Academic_Context a ON f.id = a.id
JOIN Dim_Assignments d ON f.id = d.id;

CREATE OR REPLACE VIEW gold.subject_performance_comparison AS
SELECT 
    f.Math_Score, f.Science_Score, f.Computer_Science_Score, f.English_Score, f.French_Score,
    d.assi_Math, d.assi_Science, d.assi_English, d.assi_Social_Studie, d.assi_French, d.assi_Computer_Science
FROM Fact_Performance f
JOIN Dim_Assignments d ON f.id = d.id;
```
